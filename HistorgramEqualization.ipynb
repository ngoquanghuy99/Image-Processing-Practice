```python
import cv2
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('anh4.jpg')
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
equalized_img = cv2.equalizeHist(gray_img)
```

```python
cv2.imwrite('output.png', equalized_img)
```

```python
output = cv2.imread('output.png')
plt.imshow(output)
```
