# 3. Cân bằng histogram
Cân bằng histogram để những bức ảnh rõ nét hơn. Với những bức ảnh có histogram tập trung vào một chỗ sẽ rất khó nhìn do các pixel có độ tương phản thấp. Chúng ta sẽ cân bằng histogram để chúng được dàn đều ra. Những pixel sáng thì sáng hẳn, pixel tối thì tối hẳn.

```python
import cv2
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
```

## Cân bằng histogram với ảnh xám

```python
img = cv2.imread('anh4.jpg') # đọc ảnh gốc 
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # chuyển ảnh về ảnh xám
```

```python
equalized_img = cv2.equalizeHist(gray_img) # hàm cân bằng histogram
```

```python
cv2.imwrite('output.png', equalized_img) # ghi ảnh 
```

```python

output = cv2.imread('output.png')
plt.subplot(121)
plt.title('Ảnh gốc chưa cân bằng')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.title('Ảnh sau khi cân bằng')
plt.imshow(output)
plt.show()
```

## Cân bằng histogram với ảnh màu
Với ảnh màu RGB ta sẽ đưa về HSV trước sau đó sẽ cân bằng theo V

```python
color_img = cv2.imread('anh5.bmp')
```

```python
img_yuv = cv2.cvtColor(color_img, cv2.COLOR_BGR2YUV) 
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
```

```python
img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
```

```python
plt.subplot(121)
plt.title('Ảnh gốc')
plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.title('Ảnh sau cân bằng')
plt.imshow(cv2.cvtColor(img_output, (cv2.COLOR_BGR2RGB)))
plt.show()
```

Có thể thấy, 1 bức ảnh rất tối, độ tương phản thấp, các pixel đa phần đều màu có giá trị rất thấp, ta gần như không thể nhìn rõ được các chi tiết trong vùng tối của bức ảnh. Sau khi cân bằng lược đồ histogram, trông nó "tươi sáng" hẳn lên nhờ việc tăng độ tương phản giữa các pixel có độ chênh lệnh giá trị bé.
