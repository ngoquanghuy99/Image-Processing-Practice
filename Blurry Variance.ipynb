Lần này chúng ta sẽ xác định độ nhòe của ảnh, ta sẽ chuyển ảnh về ảnh xám, sau đó dựa vào Laplacian mà tính toán phương sai để xác định độ nhòe của ảnh.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
import cv2
```

```python
def showblur(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    fm = cv2.Laplacian(gray, cv2.CV_64F).var()
    title = 'Blurry scale %f' %fm
    plt.axis('off')
    plt.title(title)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv2.imread('blur_image.jpg')
showblur(img)
```

```python
img = cv2.imread('blur_image1.jpg')
showblur(img)
```
